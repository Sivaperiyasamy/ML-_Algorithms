```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Position_Salaries.csv")
data
```

```python
x=data.iloc[:,1:2].values
x
```

```python
y=data.iloc[:,2].values
y
```

```python
#not for coing js=ust to understanding

from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x,y)
```

```python
#coding starts from here

from sklearn.preprocessing import PolynomialFeatures
poly_reg2=PolynomialFeatures(degree=2)
x_poly=poly_reg2.fit_transform(x)
lin_reg2=LinearRegression()
lin_reg2.fit(x_poly,y)
```

```python
#degree 3

from sklearn.preprocessing import PolynomialFeatures
poly_reg3=PolynomialFeatures(degree=3)
x_poly1=poly_reg3.fit_transform(x)
lin_reg3=LinearRegression()
lin_reg3.fit(x_poly1,y)
```

```python
#plotting using scatter plot

plt.scatter(x,y,color='r')
plt.plot(x,lin_reg.predict(x),color='blue') #lin_reg.predict is alrady trained model
plt.title("Linear Regression")
plt.xlabel("Position leve")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(x,y,color='r')
plt.plot(x,lin_reg2.predict(x_poly),color='b')
plt.plot(x,lin_reg3.predict(x_poly1),color='g')
plt.title("Polynomial Regression")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

```python
x=data.drop('Y house price of unit area',axis=1)
x
```

```python
y=data['Y house price of unit area']
y
```

```python
polynomial_reg=PolynomialFeatures(degree=2,include_bias=False)
```

```python
poly_feature=polynomial_reg.fit_transform(x)
poly_feature.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(poly_feature,y,test_size=0.3,random_state=101)
```

```python
 polymodel=LinearRegression()
```

```python
polymodel.fit(x_train,y_train)
```

```python
y_pred=polymodel.predict(x_test)
```

```python
pd.DataFrame({'y_test': y_test,'y_pred': y_pred,'Residual': (y_test-y_pred)})
```

```python
Poly_MAE=metrics.mean_absolute_error(y_test,y_pred)
Poly_MSE=metrics.mean_squared_error(y_test,y_pred)
Poly_RMSE=np.sqrt(Poly_MSE)
```
