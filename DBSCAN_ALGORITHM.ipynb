```python
  import pandas as pd
  import numpy as np
  import matplotlib.pyplot as plt
  from sklearn.preprocessing import StandardScaler
  from sklearn.cluster import DBSCAN
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Mall_Customers.csv")
data
```

```python
data['Gender'].replace('Male',1,inplace=True)
data['Gender'].replace('Female',2,inplace=True)
```

```python
features=['Annual Income (k$)','Spending Score (1-100)']
x=data[features]
```

```python
x_stand=StandardScaler().fit_transform(x)
x_stand
```

```python
dbscan=DBSCAN(eps=0.5,min_samples=5)
cluster=dbscan.fit_predict(x_stand)
cluster
```

```python
centroids=pd.DataFrame(x_stand)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'],c=cluster,cmap='viridis',edgecolors='k',s=50)
plt.title('DBSCAN clustring of mall coustomer data')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
