```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# # STEP 1 - INITIALISING

```python
def initialize_centroids(x,k):
  indices=np.random.choice(len(x),k,replace=True)
  return x[indices]
```

# # STEP 2 - CALCULATING DISTANCE

```python
def assign_to_clusters(x,centroids):
  distances=np.linalg.norm(x[:,np.newaxis]-centroids,axis=2)
  return np.argmin(distances,axis=1)
```

#  # mean

```python
def update_centroids(x,labels,k):
  centroids=np.array([x[labels==i].mean(axis=0) for i in range(k)])
  return centroids
```

# # calculate k means

```python
def k_mean(x,k,max_iteration=100):
  centroids=initialize_centroids(x,k)
  for _ in range(max_iteration):
    labels=assign_to_clusters(x,centroids)
    new_centroids=update_centroids(x,labels,k)

    if np.all(centroids==new_centroids):
      break
    centroids=new_centroids
  return labels,centroids
```

```python
np.random.seed(42)
x=np.concatenate([np.random.normal(loc=i,scale=1,size=(50,2)) for i in range (5)])
```

```python
x
```

```python
k=5
labels,centroids=k_mean(x,k)
```

```python
plt.scatter(x[:,0],x[:,1],c=labels,cmap='viridis')
plt.scatter(centroids[:,0],centroids[:,1],marker='x',s=200,c='red',label='centroids')
plt.title('k_means clustering')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```
